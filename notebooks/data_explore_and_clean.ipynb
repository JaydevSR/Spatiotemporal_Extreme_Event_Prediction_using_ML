```python
import re
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('..\database\spatiotemporal_trn_data.csv')
```

```python
# drop columns with all null values
df = df.dropna(axis=1, how='all')
```

```python
def get_date_dict(date_str):
    reg = re.compile(r"^(?P<yr>\d{4})-(?P<mon>\d{2})-(?P<day>\d{2})T(?P<hr>\d{2}):(?P<min>\d{2}):(?P<sec>\d{2})$")
    m = reg.match(date_str)
    return m.groupdict()
```

```python
# Seperate date data
yr = []
mon = []
day = []
hr = []
mn = []
sec = []
for idx in range(0, len(df)):
    date_dict = get_date_dict(df.DATE[idx])
    yr.append(int(date_dict['yr']))
    mon.append(int(date_dict['mon']))
    day.append(int(date_dict['day']))
    hr.append(int(date_dict['hr']))
    mn.append(int(date_dict['min']))
    sec.append(int(date_dict['sec']))
    
df['YEAR'] = pd.Series(yr)
df['MONTH'] = pd.Series(mon)
df['DAY'] = pd.Series(day)
df['HOUR'] = pd.Series(hr)
df['MINUTES'] = pd.Series(mn)
df['SECONDS'] = pd.Series(sec)
```

```python
# Add targets to same df
df['TARGETS'] = pd.read_csv('..\database\spatiotemporal_trn_targets.csv', names=['index', 'TARGETS'])['TARGETS']
```

def splt(x):
    try:
        x = float(x)
    except:
        m = re.match(r"^(?P<dig>\d+.\d+)V$", x)
        x = float(m.groupdict()['dig'])
        
    if x < 0.1:
        return 1
    elif x < 1:
        return 2
    elif x < 2:
        return 3
    elif x < 4:
        return 4
    else:
        return 5
