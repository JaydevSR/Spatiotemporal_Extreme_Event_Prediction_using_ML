```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy
import datetime
import tqdm
```

```python
def autocross(x, y):
    c = scipy.signal.correlate(x, y, mode="same")
    v = np.array([c[i]/( len(x)-abs( i - (len(x)/2)  ) ) for i in range(len(c))])
    return v
```

```python
def autocorr5(x):
    mean=x.mean()
    var=numpy.var(x)
    xp=x-mean
    corr=scipy.signal.correlate(xp,xp,mode='full')[len(x)-1:]/var/len(x)
    return np.array(corr)
```
